```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Definicja zakresu danych (z)
z = np.linspace(-5, 5, 400)

# 2. Obliczenie funkcji tanh
f_z = np.tanh(z)

# 3. Obliczenie gradientu funkcji tanh: f'(z) = 1 - tanh^2(z)
df_z = 1 - np.square(f_z)

plt.figure(figsize=(10, 6))

# Rysowanie funkcji podstawowej
plt.plot(z, f_z, label=r'$f(z) = \tanh(z)$', color='blue', linewidth=2)

# Rysowanie gradientu (pochodnej)
plt.plot(z, df_z, label=r"$f'(z) = 1 - \tanh^2(z)$", color='red', linewidth=2, linestyle='--')

plt.title('Funkcja aktywacji tanh oraz jej gradient', fontsize=16)
plt.xlabel('z', fontsize=12)
plt.ylabel('Warto', fontsize=12)
plt.axhline(0, color='black', linewidth=0.8)
plt.axvline(0, color='black', linewidth=0.8)
plt.grid(True, linestyle=':', alpha=0.6)
plt.legend(fontsize=12)

plt.show()
```
