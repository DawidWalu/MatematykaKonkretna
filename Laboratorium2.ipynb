```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img = Image.open('8.webp').convert('L')
img_np = np.asarray(img).astype(np.float64) / 255

U_row, S_row, Vt_row = np.linalg.svd(img_np, full_matrices=False)
U_col, S_col, Vt_col = np.linalg.svd(img_np.T, full_matrices=False)

corr_rows = np.corrcoef(img_np)
corr_cols = np.corrcoef(img_np.T)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
axes[0].imshow(corr_rows, cmap='hot', interpolation='nearest')
axes[0].set_title('Korelacja wierszy obrazu')
axes[1].imshow(corr_cols, cmap='hot', interpolation='nearest')
axes[1].set_title('Korelacja kolumn obrazu')
plt.tight_layout()
plt.show()
```
